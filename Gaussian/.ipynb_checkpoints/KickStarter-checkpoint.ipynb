```python
import pandas as pd
import numpy as np
import dateutil.parser as time_parser
from sklearn.neural_network import MLPClassifier
from multiprocessing import Pool
from sklearn.metrics import f1_score
from numpy.linalg import cholesky, inv

def parse(data):
    data["deadline"] = [dateutil.parser.parse(x).timestamp() for x in data["deadline"]]
    return data
```

```python
url = "../datasets/kickstarter/ks-projects-201801.csv"
X_columns = ["category", "main_category", "currency", "deadline", "backers", "country", "usd_pledged_real", "usd_goal_real"]
used_columns = X_columns + ["state"]

data = pd.read_csv(url, usecols=used_columns)
data[100:105]
```

```python
def cleansing(data):
    for col in data.select_dtypes(include=['int16', 'int32', 'int64', 'float16', 'float32', 'float64']):
        data[col] = pd.to_numeric(data[col])
        data[col] = data[col].replace(np.nan, data[col].mean())
    for col in data.select_dtypes(exclude=["number"]).columns:
        data[col] = data[col].astype('category')    
        data[col] = data[col].cat.codes
    return data
data = cleansing(data)
```

```python
X = data.loc[:, data.columns != "state"]
Y = data["state"]
amount = int(len(X) * 0.8)
Xtrain = X[0:amount]
Ytrain = Y[0:amount].astype('int')
Xtest = X[amount::]
Ytest = Y[amount::].astype('int')
Xtrain.head(10)
```

```python
mu = [Xtrain[Ytrain == i].mean() for i in range(max(Ytrain))]
sigma = [(1/(len(Xtrain[Ytrain == i])-1)) * (Xtrain[Ytrain == i].T.dot(Xtrain[Ytrain == i])) for i in range(max(Ytrain))]
mu, sigma
```

```python
#Xtests = Xtest[0]
#Ytests = Ytest[0]
Xmean = [X - mu[i] for i in range(max(Ytrain))]
z = [Xmean[i].dot(inv(cholesky(sigma[i]))) for i in range(max(Ytrain))]
[sum(item ** 2.) for item in z]
```
