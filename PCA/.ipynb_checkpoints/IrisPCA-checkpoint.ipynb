```python
url = "../../datasets/iris.data"
covariance_perc=0.74

# load dataset into Pandas DataFrame
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])

print("Non estandarized data")
df
```

```python
features = ['sepal length', 'sepal width', 'petal length', 'petal width']

# Separating out the features
x = df.loc[:, features].values

# Separating out the target
y = df.loc[:,['target']].values

# Standardizing the features
x = StandardScaler().fit_transform(x)
print("Standarized data")
prettyX=pd.DataFrame(x, columns=features)
prettyX
```

```python
sigma = ((x.transpose().dot(x))) / (len(x)-1)
sigma
```

```python
correlations = pd.DataFrame([[sigma[i, j] / (math.sqrt(sigma[i, i]) * math.sqrt(sigma[j, j])) 
                 for j in range(0, len(sigma))] for i in range(0, len(sigma))], columns=features)
print("Correlation matrix")
plt.matshow(correlations)
correlations
```

```python
#EigenV* calc
(eigenValues, eigenVectors) = np.linalg.eig(sigma)
idx = eigenValues.argsort()[::-1]   
eigenValues = eigenValues[idx]
eigenVectors = eigenVectors[:,idx]
eigenValues, eigenVectors
```

```python
#Get best K components
allLambdas = sum(eigenValues)
lambdas = [sum(eigenValues[::(len(eigenValues)-i)]) / allLambdas for i in range(0, len(eigenValues))]
plt.plot(range(1, len(eigenValues)+1), lambdas, "ro--")
plt.xlabel("Number of eigenvalues used")
plt.ylabel("Covariance")
plt.legend(["Covariance"])
plt.show()
bestEigenValues = list(filter(lambda l: l > covariance_perc, reversed(lambdas)))[0]
idx = np.where(lambdas == bestEigenValues)[0][0]
bestK = idx + 1
print('To keep {0} of the variance, at least {1} of the {2} eigen* components MUST be used.\n'.format(covariance_perc, bestK, len(lambdas)))
      
```

```python
bestEigenValues = eigenValues[0:bestK]
bestEigenVectors = eigenVectors[0:bestK]

print(bestEigenVectors)
print(bestEigenValues)
```

```python
reconstruct = bestEigenVectors.transpose().dot(x.transpose()).transpose()
pd.DataFrame(reconstruct)
```

```python
xrec = pd.DataFrame(bestEigenVectors.dot(reconstruct.transpose()).transpose())
xrec
```
