# Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
url = "../datasets/car/imports-85.data"

# load dataset into Pandas DataFrame
df = pd.read_csv(url, na_values= ['?'])
features = df.columns

print("Non estandarized data")
df.head()
```

```python
for col in df.select_dtypes(include=['int16', 'int32', 'int64', 'float16', 'float32', 'float64']):
    df[col] = pd.to_numeric(df[col])
    df[col] = df[col].replace(np.nan, df[col].mean())
```

```python
for col in df.select_dtypes(exclude=["number"]).columns:
    df[col] = df[col].astype('category')    
    df[col] = df[col].cat.codes
```

```python
df.to_csv("../datasets/car/imports-85-clean.data")
```

# Standarization

```python
from sklearn.preprocessing import StandardScaler
```

```python
list(features[:-1])
```

```python
x = df.loc[:, df.columns != 'price'].values
y = df.loc[:, ['price']].values

x = StandardScaler().fit_transform(x)
xPretty = pd.DataFrame(x, columns=list(features[:-1]))
xPretty.head()
```

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_predict

lr = linear_model.LinearRegression()

# cross_val_predict returns an array of the same size as `y` where each entry
# is a prediction obtained by cross validation:
predicted = cross_val_predict(lr, x, y, cv=10)

"""fig, ax = plt.subplots()
ax.scatter(y, predicted, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()"""
```
